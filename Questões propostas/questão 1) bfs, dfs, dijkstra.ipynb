```python
def bfs_shortest_path(graph, start, end):

  queue = deque([(start, 0)]) # Fila para BFS, armazena (nó, distância)
  visited = {start} # Conjunto de nós visitados

  while queue:
    node, distance = queue.popleft()

    if node == end:
      return distance

# Iterando sobre o nós (percorrendo os vizinhos)
    for neighbor in graph[node]:
      if neighbor not in visited:
        visited.add(neighbor) # Marca como visitado
        queue.append((neighbor, distance + 1)) # Incrementa a distância para o vizinho

  return float('inf') # Se o nó final não for encontrado

  # Dicionário com as estações de metrô de Recife

if __name__ == "__main__":
  graph = defaultdict(dict)

  # Linha Centro-1:
  graph['Jaboatão']['Engenho Velho'] = 5
  graph['Engenho Velho']['Floriano'] = 5
  graph['Floriano']['Cavaleiro'] = 12
  graph['Cavaleiro']['Coqueiral'] = 4
  graph['Coqueiral']['Tejipió'] = 4
  graph['Tejipió']['Barro'] = 4
  graph['Barro']['Werneck'] = 2
  graph['Werneck']['Santa Luzia'] = 3
  graph['Santa Luzia']['Mangueira'] = 4
  graph['Mangueira']['Ipiranga'] = 2
  graph['Ipiranga']['Afogados'] = 5
  graph['Afogados']['Joana Bezerra'] = 6

  # Linha Centro-2:
  graph['Camaragibe']['Cosme e Damião'] = 6
  graph['Cosme e Damião']['Rodoviária'] = 5
  graph['Rodoviária']['Curado'] = 5
  graph['Curado']['Alto do Céu'] = 3
  graph['Alto do Céu']['Coqueiral'] = 6
  graph['Coqueiral']['Tejipió'] = 4
  graph['Tejipió']['Barro'] = 4
  graph['Barro']['Werneck'] = 2
  graph['Werneck']['Santa Luzia'] = 3
  graph['Santa Luzia']['Mangueira'] = 4
  graph['Mangueira']['Ipiranga'] = 2
  graph['Ipiranga']['Afogados'] = 5
  graph['Afogados']['Joana Bezerra'] = 6

  # Linha Sul:
  graph['Recife']['Joana Bezerra'] = 5
  graph['Joana Bezerra']['Largo da Paz'] = 5
  graph['Largo da Paz']['Imbiribeira'] = 4
  graph['Imbiribeira']['Antônio Falcão'] = 10
  graph['Antônio Falcão']['Shopping'] = 3
  graph['Shopping']['Tancredo'] = 3
  graph['Tancredo']['Aeroporto'] = 6
  graph['Aeroporto']['Porta Larga'] = 6
  graph['Porta Larga']['Monte dos Guararapes'] = 4
  graph['Monte dos Guararapes']['Prazeres'] = 4
  graph['Prazeres']['Cajueiro Seco'] = 5
  start_node = 'Barro'
  end_node = 'Tancredo'

  shortest_distance = bfs_shortest_path(graph, start_node, end_node)

if shortest_distance == float('inf'):
  print(f'Não foi encontrado caminho entre {start_node} e {end_node}.')
else:
  print(f'Distância mais curta entre {start_node} e {end_node} levará {shortest_distance} estações.')

start_time = time.time()
shortest_distance = bfs_shortest_path(graph, start_node, end_node)
end_time = time.time()

execution_time = end_time - start_time

print(f'\nTempo de execução da função BFS: {execution_time} segundos')
```

# Algoritmo de Dijkstra


```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict
import time
```

A função dijkstra implementa o algoritmo de Dijkstra para encontrar o caminho mais curto entre dois nós (start e end) em um grafo ponderado (graph).

```python
def dijkstra(graph, start, end):
  # Cria um conjunto com todos os nós do grafo.
  all_nodes = set(graph.keys())
  for neighbors in graph.values():
    all_nodes.update(neighbors.keys()) # Use update to add elements to the set.
  distances = {node: float('inf') for node in all_nodes}
  distances[start] = 0
  # Cria uma fila de prioridade queue para armazenar os nós a serem visitados
  queue = [(0, start)]
  # Cria um dicionário previous_nodes para armazenar o nó anterior (pai) no caminho mais curto para cada nó.
  previous_nodes = {}

  # Enquanto a fila de prioridade possuir nós, salve a distância e o nós atual nas variáveis
  while queue:
    current_distance, current_node = heapq.heappop(queue)

    if current_distance > distances[current_node]:
      continue

    if current_node == end:
      break

    # Check if current_node has neighbors before iterating
    if current_node in graph:
      for neighbor, weight in graph[current_node].items():
        distance = current_distance + weight

        if distance < distances[neighbor]:
          distances[neighbor] = distance
          previous_nodes[neighbor] = current_node
          heapq.heappush(queue, (distance, neighbor))

  # Check if a path was found before trying to reconstruct it
  if end in previous_nodes:
    path = []
    node = end
    while node != start:
      path.append(node)
      node = previous_nodes[node]
    path.append(start)
    path.reverse()

    return distances[end], path
  else:
    # Return infinity and an empty path if no path is found
    return float('inf'), []

def add_edge(graph, u, v, weight):
    graph[u][v] = weight
    graph[v][u] = weight  # Adiciona a aresta no sentido inverso

# Dicionário com as estações de metrô de Recife

if __name__ == "__main__":
  graph = defaultdict(dict)

  # Linha Centro-1:
  graph['Jaboatão']['Engenho Velho'] = 5
  graph['Engenho Velho']['Floriano'] = 5
  graph['Floriano']['Cavaleiro'] = 12
  graph['Cavaleiro']['Coqueiral'] = 4
  graph['Coqueiral']['Tejipió'] = 4
  graph['Tejipió']['Barro'] = 4
  graph['Barro']['Werneck'] = 2
  graph['Werneck']['Santa Luzia'] = 3
  graph['Santa Luzia']['Mangueira'] = 4
  graph['Mangueira']['Ipiranga'] = 2
  graph['Ipiranga']['Afogados'] = 5
  graph['Afogados']['Joana Bezerra'] = 6

  # Linha Centro-2:
  graph['Camaragibe']['Cosme e Damião'] = 6
  graph['Cosme e Damião']['Rodoviária'] = 5
  graph['Rodoviária']['Curado'] = 5
  graph['Curado']['Alto do Céu'] = 3
  graph['Alto do Céu']['Coqueiral'] = 6
  graph['Coqueiral']['Tejipió'] = 4
  graph['Tejipió']['Barro'] = 4
  graph['Barro']['Werneck'] = 2
  graph['Werneck']['Santa Luzia'] = 3
  graph['Santa Luzia']['Mangueira'] = 4
  graph['Mangueira']['Ipiranga'] = 2
  graph['Ipiranga']['Afogados'] = 5
  graph['Afogados']['Joana Bezerra'] = 6

  # Linha Sul:
  graph['Recife']['Joana Bezerra'] = 5
  graph['Joana Bezerra']['Largo da Paz'] = 5
  graph['Largo da Paz']['Imbiribeira'] = 4
  graph['Imbiribeira']['Antônio Falcão'] = 10
  graph['Antônio Falcão']['Shopping'] = 3
  graph['Shopping']['Tancredo'] = 3
  graph['Tancredo']['Aeroporto'] = 6
  graph['Aeroporto']['Porta Larga'] = 6
  graph['Porta Larga']['Monte dos Guararapes'] = 4
  graph['Monte dos Guararapes']['Prazeres'] = 4
  graph['Prazeres']['Cajueiro Seco'] = 5

  # Tornando o grafo bidirecional
  add_edge(graph, 'Jaboatão', 'Engenho Velho', 5)
  add_edge(graph, 'Engenho Velho', 'Floriano', 5)
  add_edge(graph, 'Floriano', 'Cavaleiro', 12)
  add_edge(graph, 'Cavaleiro', 'Coqueiral', 4)
  add_edge(graph, 'Coqueiral', 'Tejipió', 4)
  add_edge(graph, 'Tejipió', 'Barro', 4)
  add_edge(graph, 'Barro', 'Werneck', 2)
  add_edge(graph, 'Werneck', 'Santa Luzia', 3)
  add_edge(graph, 'Santa Luzia', 'Mangueira', 4)
  add_edge(graph, 'Mangueira', 'Ipiranga', 2)
  add_edge(graph, 'Ipiranga', 'Afogados', 5)
  add_edge(graph, 'Afogados', 'Joana Bezerra', 6)
  add_edge(graph, 'Camaragibe', 'Cosme e Damião', 6)
  add_edge(graph, 'Cosme e Damião', 'Rodoviária', 5)
  add_edge(graph, 'Rodoviária', 'Curado', 5)
  add_edge(graph, 'Curado', 'Alto do Céu', 3)
  add_edge(graph, 'Alto do Céu', 'Coqueiral', 6)
  add_edge(graph, 'Coqueiral', 'Tejipió', 4)
  add_edge(graph, 'Tejipió', 'Barro', 4)
  add_edge(graph, 'Barro', 'Werneck', 2)
  add_edge(graph, 'Werneck', 'Santa Luzia', 3)
  add_edge(graph, 'Santa Luzia', 'Mangueira', 4)
  add_edge(graph, 'Mangueira', 'Ipiranga', 2)
  add_edge(graph, 'Ipiranga', 'Afogados', 5)
  add_edge(graph, 'Afogados', 'Joana Bezerra', 6)
  add_edge(graph, 'Recife', 'Joana Bezerra', 5)
  add_edge(graph, 'Joana Bezerra', 'Largo da Paz', 5)
  add_edge(graph, 'Largo da Paz', 'Imbiribeira', 4)
  add_edge(graph, 'Imbiribeira', 'Antônio Falcão', 10)
  add_edge(graph, 'Antônio Falcão', 'Shopping', 3)
  add_edge(graph, 'Shopping', 'Tancredo', 3)
  add_edge(graph, 'Tancredo', 'Aeroporto', 6)
  add_edge(graph, 'Aeroporto', 'Porta Larga', 6)
  add_edge(graph, 'Porta Larga', 'Monte dos Guararapes', 4)
  add_edge(graph, 'Monte dos Guararapes', 'Prazeres', 4)
  add_edge(graph, 'Prazeres', 'Cajueiro Seco', 5)

  # Seleção de estação de partida e chegada
  start_node = 'Jaboatão'
  end_node = 'Cajueiro Seco'

  shortest_distance, shortest_path = dijkstra(graph, start_node, end_node)
  if shortest_distance == float('inf'):
    print(f'Não foi encontrado caminho entre {start_node} e {end_node}.')
  else:
    print(f'Distância mais curta entre {start_node} e {end_node} levará {shortest_distance} minutos.')
    print(f'Melhor rota: {shortest_path}.')

  # Avaliando o tempo computacional para o Dijkstra
  start_time = time.time()
  shortest_distance, shortest_path = dijkstra(graph, start_node, end_node)
  end_time = time.time()

  execution_time = end_time - start_time

  print(f'\nTempo de execução da função Dijkstra: {execution_time} segundos')
```
